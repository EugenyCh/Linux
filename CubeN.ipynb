```python
import numpy as np
```

```python
from copy import deepcopy
```

```python
N = 6
```

```python
vers = []
```

```python
verOne = np.ones(shape=(N,))
```

```python
faceOne = [deepcopy(verOne) for i in range(4)]
```

```python
ids = list(range(N))
```

```python
import itertools
```

```python
cmb = list(itertools.combinations(ids, 2))
cmb
```

```python
for p in cmb:
    a = p[0]
    b = p[1]
    otherIds = [c for c in ids if c not in p]
    print(p)
    face = deepcopy(faceOne)
    face[1][a] = -1.0
    face[2][a] = -1.0
    face[2][b] = -1.0
    face[3][b] = -1.0
    print(face)
    otherCmb = []
    for n in range(1, N - 1):
        otherCmb.extend(itertools.combinations(otherIds, n))
    vers.extend(face)
    for s in otherCmb:
        faceN = deepcopy(face)
        for i in s:
            for j in range(4):
                faceN[j][i] = -1.0
        vers.extend(faceN)
```

```python
len(vers)
```

```python
for i in range(len(vers)):
    vers[i] = vers[i].tolist()
```

```python
import json
```

```python
polygons = []
```

```python
for i in range(len(vers) // 4):
    polygons.append(vers[i * 4:i * 4 + 4])
```

```python
len(polygons)
```

```python
data = {
    "Dimension": N,
    "Polygons": polygons
}
```

```python
with open("Cube" + str(N) + ".json", "w") as file:
    json.dump(data, file)
```
